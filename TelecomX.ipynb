## CARGUE Y AJUSTE DE ARCHIVO .JSON

```python
%pip install summarytools -q
import pandas as pd
import requests
import seaborn as sns
from summarytools import dfSummary
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json'
response = requests.get(url)

if response.status_code == 200:
    data_json = response.json()
else:
    print("Error al descargar el archivo:", response.status_code)
```

```python
df = pd.DataFrame(data_json)
```

```python
df.head(5)
```

```python
cols_base = ['customerID', 'Churn']

df_customer = pd.json_normalize(df['customer'])
df_phone    = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account  = pd.json_normalize(df['account'])

df_v1 = pd.concat([df[cols_base], df_customer, df_phone, df_internet, df_account], axis=1)

df_v1.head()
```

```python
df_v1.columns
```

#Diccionario de datos

* customerID: número de identificación único de cada cliente
* Churn: si el cliente dejó o no la empresa
* gender: género (masculino y femenino)
* SeniorCitizen: información sobre si un cliente tiene o no una edad igual o mayor a 65 años
* Partner: si el cliente tiene o no una pareja
* Dependents: si el cliente tiene o no dependientes
* tenure: meses de contrato del cliente
* PhoneService: suscripción al servicio telefónico
* MultipleLines: suscripción a más de una línea telefónica
* InternetService: suscripción a un proveedor de internet
* OnlineSecurity: suscripción adicional de seguridad en línea
* OnlineBackup: suscripción adicional de respaldo en línea
* DeviceProtection: suscripción adicional de protección del dispositivo
* TechSupport: suscripción adicional de soporte técnico, menor tiempo de espera
* StreamingTV: suscripción de televisión por cable
* StreamingMovies: suscripción de streaming de películas
* Contract: tipo de contrato
* PaperlessBilling: si el cliente prefiere recibir la factura en línea
* PaymentMethod: forma de pago
* Charges.Monthly: total de todos los servicios del cliente por mes
* Charges.Total: total gastado por el cliente

```python
# Considero que para este analisis y teniendo en cuenta que los valores vacios de "Churn" Corresponden al "3%"
# del total de datos podriamos eliminarlos, ya que esta es la variable mas importante y no pódriamos
# continuar con este valor vacio, se podria usar tenicas mas avanzas para inferir si continua o no con
# el servicio usando machine learning pero no aplica para ese primer analisis.
```

```python
df_v2 = df_v1[df_v1['Churn'] != '']
```

### Corrigiendo valores nulos "Charges.Total"

```python
num_vacios = (df_v2['Charges.Total'] == ' ').sum() # aqui se identico los valores vacios al tanteo ya que no es " " con un espacio sino "  " con dos espacios

num_no_vacios = (df_v2['Charges.Total'] != ' ').sum()

print(f"Valores vacíos (''): {num_vacios}")
print(f"Valores no vacíos: {num_no_vacios}")
```

```python
# Como en la conclusion anterior podemos escoger eliminar los valores o para este caso he decidido inferir el valor con el producto entre pago mensual y antiguedad
```

```python
df_v2.loc[:, 'Charges.Total'] = df_v2['Charges.Total'].replace(" ", np.nan)
```

```python
estimador = (df_v2['Charges.Monthly'] * df_v2['tenure'].clip(lower=1)).where(df_v2['Charges.Total'].isna()) # el metodo clip lo investigue es 1 si #tenure# es menor que 1,
df_v2.loc[:, 'Charges.Total'] = df_v2['Charges.Total'].fillna(estimador)                                      # esto porque hay caso en que el afiliado este empezando el primer mes
```

```python
df_v3 = df_v2.copy()
df_v3['Charges.Total'] = pd.to_numeric(df_v3['Charges.Total'], errors='coerce')
```

### Establecer columnas booleanas

```python
df_v3.head()
```

```python
cols_booleanas = ["Churn","Dependents", "PaperlessBilling", "SeniorCitizen", "PhoneService","Partner"]


def boolean(val):
    if str(val).strip().lower() in ['yes', '1', 'true']:
        return True
    elif str(val).strip().lower() in ['no', '0', 'false']:
        return False
    else:
        return pd.NA


for col in cols_booleanas:
    df_v3[col] = df_v3[col].apply(boolean).astype('boolean')
```

```python
df_v3.dtypes
```

# Analisis y correlaciones

Con respecto a los cargos mensuales y cargos totales, estos no presentan mayor diferencia especialmente los cargos mensuales es decir los que se van y los que se quedan pagan valores similares mensualmente sin embargo la **antiguedad(Tenure)** si representa una diferencia significativa: *afiliaciones mas recientes son mas propensas a abandonar las suscripciones*  



```python
df_v3.groupby('Churn')[["Charges.Total","Charges.Monthly","tenure"]].mean().T.sort_values(by=True, ascending=False).round(1)
```

Con respeto a las variables boleanas de la base de datos ninguna representa un insight signifcativo, acontinuacion se presenta un arbol de decisiones y su proporción de los clientes que abandonaron la compañia, de lo cual se puede resumir que el **perfil** del cliente que abandona la compañia con un 30% de regularia es aquel que: *Tiene servicio de telefono, no tiene dependiente ni pareja, prefiere recibir la fatura en linea y es menor de 65 años*

```python
df_churn_yes = df_v3[df_v3['Churn'] == True]

df_churn_yes[["PhoneService","Dependents","Partner","PaperlessBilling", "SeniorCitizen"]].value_counts(normalize=True)
```

Con el resto de informacion se puede **concluir**:



*   Con respeto al *genero y las suscripcion a mas de una linea telefonica* las proporsiones son similares  
*   El *tipo de internet* si influye el 70% de las decersiones de dieron cuando el servicio se prestaba por fibra optica
* El 78% de los clientes que no tenian *OnlineSecurity* y el 66% que no contaban con el servicio de *OnlineBackup* dejaron la compañia
* El 65% de los usuarios que no contaban con *DeviceProtection* abandonaron la compañia
* Los servicios de streaming no son relevantes en la permanecia de los usuarios
* Los usuarios que tenian contrato tipo: **Month-to-month** fueron un 89% mas propenson a abandonar la compañia al igual que los no contaban con soporte tecnico en un 78%
   



```python
cols = ["gender", "MultipleLines", "InternetService", "OnlineSecurity","OnlineBackup","DeviceProtection","StreamingTV",
        "StreamingMovies", "Contract", "PaymentMethod", "TechSupport"]

for col in cols:
    print(f"\n🔍 Distribución de '{col}' según Churn:\n")
    result = df_v3.groupby('Churn')[col].value_counts(normalize=True).unstack().T
    print(result.round(3))
```
